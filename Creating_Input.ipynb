```python
import cv2
import mediapipe as mp
import csv
import numpy as np
import pandas as pd
import os
import pickle
```

```python
def Video_analyse(video_path):
    mp_drawing = mp.solutions.drawing_utils
    mp_drawing_styles = mp.solutions.drawing_styles
    mp_pose = mp.solutions.pose

    cap = cv2.VideoCapture(video_path)
    number_of_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    print(f"Number of frames for {video_path}: {number_of_frames} and FPS is {fps}")

    x = []
    y = []
    length = []

    with mp_pose.Pose(
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5) as pose:
        while cap.isOpened():
            success, image = cap.read()
            if not success:
                print("Ignoring empty camera frame.")
                break

            image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
            image.flags.writeable = False
            results = pose.process(image)

            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
            mp_drawing.draw_landmarks(
                image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
            if not results.pose_landmarks:
              continue
            for idx,landmark in enumerate(results.pose_landmarks.landmark):
                x.append(landmark.x)
                y.append(landmark.y)
                #length.append(landmark)

    cap.release()
    cv2.destroyAllWindows()
    length = x
    length.extend(y)
    return length,number_of_frames
```

```python
file_name_train = "/content/drive/MyDrive/Capstone/train.pkl"
file_name_target = "/content/drive/MyDrive/Capstone/target.pkl"
file_name_index = "/content/drive/MyDrive/Capstone/index.pkl"
```

```python
directory = '/content/drive/MyDrive/Capstone/EVERYDAY TERMS - 1300'
c = 0
d = 0
target=[]
final = []
final_index = []

for dirname in os.listdir(directory):
  new_dir = directory + '/' + dirname
  if os.path.isdir(new_dir):
    #print(new_dir)
    for folder in os.listdir(new_dir):
      path = new_dir + '/' + folder
      #print(path)
      #break
      if os.path.isdir(path):
        for file in os.listdir(path):
          new_file = path + '/' + file
          out,frame_count = Video_analyse(new_file)
          if len(out) != (frame_count*66):
            d+=1
            print(d)
            continue
          out_arr = np.array(out).reshape(frame_count,66)
          print(out_arr.shape)
          final.append(out_arr)
          #print(new_file)
          #break
          file = file.replace('.mp4','')
          #print(file)
          target.append(file)
          index = []
          idx = target.index(file)
          for i in range(frame_count):
            if i < 24:
              index.append(0)
            else:
              index.append(idx)
          index_arr = np.array(index)
          final_index.append(index_arr)
          c+=1
          print(c)
      else:
        out,frame_count = Video_analyse(path)
        if len(out) != (frame_count*66):
          d+=1
          print(d)
          continue
        out_arr = np.array(out).reshape(frame_count,66)
        print(out_arr.shape)
        final.append(out_arr)
        folder = folder.replace('.mp4','')
          #print(path)
        #print(folder)
        target.append(folder)
        index = []
        idx = target.index(folder)
        for i in range(frame_count):
          if i <24:
            index.append(0)
          else:
            index.append(idx)
        index_arr = np.array(index)
        final_index.append(index_arr)
        c+=1
        print(c)

        if c%100 == 0:
          with open(file_name_train,'ab+') as file:
            pickle.dump(final,file)
          with open(file_name_target,'ab+') as file:
            pickle.dump(target,file)
          with open(file_name_index,'ab+') as file:
            pickle.dump(final_index,file)

          final = []
          target = []
          final_index = []

print(f"Total numer of videos processed: {c} and number of skipped {d}")
```
